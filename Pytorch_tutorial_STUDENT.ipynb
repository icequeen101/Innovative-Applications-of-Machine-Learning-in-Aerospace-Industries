# **1. Comments**

When coding, it's helpful to use comments in your code. \\
This helps you, as well as others, understand how/why you wrote the code the way that you did. \\
In Python, you can make a comment in two different ways:

1.   You can use a # before your comment. \\
This is called an *in-line* comment because everything in the same line after the # will be considered a comment, but the next line returns to normal code:
      ```
      # This is a comment
      This is not a comment
      ```


2.   You can also use triple quotes around your comment. This is called a *multi-line* comment because all of the text between the triple quotes at the beginning and end are a part of the comment. These are commonly used at the beginning of functions (discussed later) to list out all of the variables that are input into the function.
      ```
      '''
      This is a comment
      This is also a comment
      '''
      ```

# **5. For Loops & Attributes**


When we want to "loop" through a list of values we use a **for each** loop.

```
# Declare a variable named devices and set it equal to a list of strings
devices = ["cuda", "mps", "cpu"]

# Declare a for each loop using a temporary variable named z
for z in devices:
    print(z)
```

In the example above the z variable gets set equal to each object in the list, one at a time. \\
This is so that we can use each value to run commmands, like *print statements*. \\
The output to the above example is:
```
cuda
mps
cpu
```

When we want to "loop" through a pre-determined range we use a **for in range** loop. \\
This loop iterates over a sequential series of numbers, starting at 0, generated by the `range()` function.

```
# Declare a for in range loop using a temporary variable named k
for k in range(3):
    print(k)
```

The output to the above example is:
```
0
1
2
```

NOTE: Every `for` loop must end with a `:` and the code inside the loop must be indented!
The code for the loop ends once a line of code is no longer indented. \\
